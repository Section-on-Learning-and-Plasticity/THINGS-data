# Working with ROIs

Often times, you might want to limit your analyses to specific regions of interest in the brain, either to study different anatomical locations separately from one another, or to simply save computational resources.

The THINGS-fMRI dataset comes with plenty localized ROIs. These ROIs include functionally defined regions that respond selectively to specific object categories (e.g. fusiform face area, extrastriate body part area, etc.). They also include retinotopically defined regions of interest like V1, V2, hV4, etc. Finally, the data also comes with estimated labels from the comprehensive [Glasser parcellation atlas](https://doi.org/10.1038/nature18933).

# Load the data

All this information is part of the `VoxelMetadata` which you can [download alongside the single trial fMRI responses from figshare](https://doi.org/10.25452/figshare.plus.20492835.v2). Let's have a look!

```python
# We'll use the data loader that comes with the python module contained in this github repo, but you may also simply load the csv files on your own as needed.
dl = ThingsmriLoader(
    thingsmri_dir='/Users/olivercontier/bigfri/openneuro/THINGS-data/THINGS-fMRI/derivatives'
)

# we're gonna chose the first subject
sub = '01'

# and load the data
responses, stimdata, voxdata = dl.load_responses(sub)
```

Let's have a look at the `VoxelMetadata`. As you can see, there is a lot of information included here, like noise ceiling estimates (`nc_singletrial` and `nc_testset`) or pRF parameter estimates (`prf-polarangle` or `prf-eccentricity`). But, we also have a lot of ROis encoded:

```python
# each row corresponds to a voxel
# each column to a metainfo paramter
voxdata
```

```python
print(*voxdata.columns, sep='\n')
```

# Extract data from an ROI

```python
# Say we are interested in V1
ROI_select = 'V1'

# We simply define a mask of all voxels that match this ROI
roimask = voxdata[ROI_select].values.astype(bool)

# and apply it to the response data
roidata = responses[roimask]

# The result will be a data frame where each row contains a voxel in the target ROI
roidata
```

```python
# let's inspect the mean V1 response over the first 100 images
plt.plot(range(100), roidata.mean(axis=0).to_numpy()[:100])
plt.xlabel('trial number');plt.ylabel('mean V1 amplitude')
plt.show()
```

# Select multiple ROIs at once

```python
# Let's say, you want to extract data from IT, which should include the following glasser labels
IT_rois = [
    'glasser-TE1p','glasser-TE2p', 'glasser-FFC', 'glasser-VVC', 'glasser-VMV2', 
    'glasser-VMV3', 'glasser-PHA1', 'glasser-PHA2', 'glasser-PHA3'
    ]

# we can select all columns in the voxel metadata
# and sum over them to create a boolean mask that reflects their union
roimask = voxdata[IT_rois].sum(axis=1).values.astype(bool)

# again, we can apply this mask to the response data to get the activation of IT voxels to each image
roidata = responses[roimask]
```

```python
# Finally, let's have a look at the data from two voxels in this IT mask
# for the first 100 images
plt.plot(range(100), roidata.iloc[1, :100].to_numpy())
plt.plot(range(100), roidata.iloc[2, :100].to_numpy())
plt.legend(['IT voxel 1', 'IT voxel 2'])
plt.xlabel('trial number');plt.ylabel('voxel amplitude')
plt.show()
```

```python
# We can also select other voxel-wise information in this region of interest,
# Like the noise ceilings estimated based on the test dataset
IT_nc = voxdata.loc[roimask, 'nc_testset'].to_numpy()

plt.hist(IT_nc, bins=50, log=True)
plt.title(f'noise ceilings in IT, subject {sub}')
plt.xlabel('Noise ceiling [%]');plt.ylabel('voxel count [log]')
plt.show()
```

# Converting back to volumetric space

Once we have analzed our ROI data, we may want to transform the result back into the 3D volumetric brain space. For that, we can use the [brainmasks which are available on figshare](https://doi.org/10.25452/figshare.plus.20492757.v1).

```python
!ls /Users/olivercontier/bigfri/openneuro/THINGS-data/THINGS-fMRI/derivatives/brainmasks
```

```python
# If you've saved the brainmasks in the same root directory as the response data, under `brainmasks/`, 
# you can use the data loader. Otherwise, you can simply define the path to the brainmask yourself
brainmask = dl.get_brainmask(sub)
```

```python
# Let's say these are some statistical results we've obtained from analyzing our ROI.
# We'll just generate some random data here
results = np.random.randn(roidata.shape[0])

# First, create a placeholder array that has the correct shape
nvoxels_brain = responses.shape[0]
results_wholebrain = np.zeros(nvoxels_brain)

# populate the whole brain array with the results from the ROI
results_wholebrain[roimask] = results

# we can use nilearn's unmask function to create a 3D image from the spatially unstructured array
results_img = unmask(results_wholebrain, brainmask)

# and save the file to disk
results_img.to_filename(f'results_ROI_demo.nii.gz')
```
